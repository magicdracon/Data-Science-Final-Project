### Background And Motivation

**“There are two things that are important in politics. The first is money, and I can’t remember what the second one is.”** ― Mark Hanna, U.S. Senator from Ohio and former chairman of the Republican National Committee

Recently, political discourse has become increasingly heated with an increasing amount of money being used to influence elections. According to __[Anna Massoglia from the Brennan Center](https://www.brennancenter.org/our-work/research-reports/dark-money-hit-record-high-19-billion-2024-federal-races)__ in 2025, Super PACs spending totaled $1.9 billion for the 2024 election cycle. A Super PAC is a committee that is allowed to spend unlimited funding support or opposing candidates as long as they don't directly work with the candidates. Traditionally, PAC spending was limited to $5000 per candidate. However, in the 2010 Supreme Court case Citizens United v. The Federal Election Commission, a decision was made that donations from PACs over $5000 are allowed if they do not go directly to the intended candidate. This carved out the way for Super PAsS where donations can be as large as possible but the money has to be used independently from the candidate. This leads us to the project, where we want to analyze these Super PACs with publicly available data from the FEC.

### Data Cleaning and Processing
As mentioned above, in Data Description and Acquisition, each item in our dataset has 24 different attributes, many of which are repetitive, extremely unlikely to be useful for our purposes, or used inconsistently enough to be effectively useless in analysis. 


In choosing which categories to keep while simplifying in our data (in total, 12 of the 24 original), we made sure to allow ourselves plenty of room for choice later on in our project.  For example, out of the many categories related to  location, we kept state and congressional districts because we believe that this category might be uniquely useful, when describing location. It might be interesting, for example, to see how often candidates for senate receive out-of-state donations. However, that would require getting more data about the PACs as the current dataset does not have that.  The congressional districts can help us understand how swing districts are possibly influenced by these donations.


For data cleaning and conversion, initially, we won’t have to do much.  The biggest problems will be the Expenditure Date category, which is in a mm/dd/year string format (we will likely convert this to a long in the form of yearmmdd, and the Payee Name and Candidate Name categories, which we know will be repeated regularly throughout the data, but may contain differences in capitalization.  We will solve this by changing each name to be all capital letters.  The other categories, including those with strings that have been standardized like the State categories and Candidate Party, will not require immediate cleaning.


There are many other categories which will require further conversion between data types to be as useful and as easy to work with as possible, especially those containing quantitative data in the form of strings.  We plan to convert the Support/Oppose Indicator category from string (where every item has either an “S” for Support or “O” for Oppose) to boolean, where the value is true if a donation is in support of a candidate, and false if it is in opposition.  We have also discussed potentially creating two new categories by separating the first letter from the following numbers in the category Election Type (the first letter represents a type of election, ie G = general, R = runoff, etc., while the following numbers represent the year of an election). 


This dataset contains largely non-quantitative data, which means that we will be doing a lot of grouping and counting of individual categories in order to meaningfully interpret and present results.
